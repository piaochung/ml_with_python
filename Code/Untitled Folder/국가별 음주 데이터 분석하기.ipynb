```python
file_path = './data/drinks.csv'
drinks_df = pd.read_csv(file_path)
drinks_df.tail()
```

```python
corr = drinks_df[['beer_servings', 'wine_servings']].corr(method = 'pearson')
corr
```

```python
%matplotlib inline
cols_view = ['beer', 'spirit', 'wine', 'alcohol']
sns.set(font_scale = 1.5)
hm = sns.heatmap(corr.values,
                cbar = True,
                annot = True,
                square = True,
                fmt = '.2f',
                annot_kws={'size': 15},
                yticklabels=cols_view,
                xticklabels=cols_view
                )
plt.tight_layout()
plt.show()

sns.set(style='whitegrid', context='notebook')
sns.pairplot(drinks_df[['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']], height=2.5)
plt.show()
```
