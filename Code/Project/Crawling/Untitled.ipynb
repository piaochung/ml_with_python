```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
base_url = 'https://movie.naver.com'
source_url = '/movie/sdb/rank/rmovie.nhn?sel=cur&date=20210222'
```

```python
req = requests.get(base_url + source_url)
html = req.content
soup = BeautifulSoup(html, 'lxml')
contents_table = soup.find(name='table')
table_body = contents_table.find(name='tbody')
table_row = table_body.find_all(name='tr')
```

```python
movie_data = []
for idx in range(1, 11):
    td = table_row[idx].find_all('div', 'tit5')
    td_base = td[0].find_all(name='a')
    td_url = base_url + td_base[0].get('href').replace('basic', 'point')
    td_title = td_base[0].get('title')
    movie_data.append([td_url, td_title])
```

```python
review_url = []

for idx in range(0, len(movie_data)):
    req = requests.get(movie_data[idx][0])
    html = req.content
    soup = BeautifulSoup(html, 'lxml')
    contents_table = soup.find('div', 'obj_section')
    table_row = contents_table.find_all('iframe')
    review_url.append(base_url + table_row[0].get('src'))
```

```python
review_text = []

for idx in range(0, len(review_url)):
    req = requests.get(review_url[idx])
    html = req.content
    soup = BeautifulSoup(html, 'lxml')
    for i in range(500):
        contents_table = soup.find('div', 'score_result')
        table_body = contents_table.find(name='ul')
        score = table_body.find_all('div', 'star_score')[0]
        print(score)
        score = score.find_all(name='em')
        review = table_body.find_all('span', id='_filtered_ment_0')
        review_text.append([score.text, review.text])
```
